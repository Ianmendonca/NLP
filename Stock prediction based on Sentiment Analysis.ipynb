```python
import pandas as pd
```

```python
df=pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

## divide the data into train and test


```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

## removing any special characters from the data

```python
# Removing punctuations
dataTrain=train.iloc[:,2:27]# all rows and all columns except frist two
dataTrain.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_IndexTrain=[str(i) for i in list1]
dataTrain.columns= new_IndexTrain
dataTrain.head(5)
```

```python
# Removing punctuations
dataTest=test.iloc[:,2:27]# all rows and all columns except frist two
dataTest.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_IndexTest=[str(i) for i in list1]
dataTest.columns= new_IndexTest
dataTest.head(5)
```

## converting all to lower case

```python
# Convertng headlines to lower case
for index in new_IndexTrain:
    dataTrain[index]=dataTrain[index].str.lower()
dataTrain.head(1)
```

```python
# Convertng headlines to lower case
for index in new_IndexTest:
    dataTest[index]=dataTest[index].str.lower()
dataTest.head(1)
```

```python
' '.join(str(x) for x in dataTrain.iloc[1,0:25])# joining all the comments
```

```python
headlinesTrain = []
for row in range(0,len(dataTrain.index)):
    headlinesTrain.append(' '.join(str(x) for x in dataTrain.iloc[row,0:25]))
```

```python
headlinesTrain[0]
```

```python
headlinesTest = []
for row in range(0,len(dataTest.index)):
    headlinesTest.append(' '.join(str(x) for x in dataTest.iloc[row,0:25]))
headlinesTest[0]
```

```python
headlinesTest[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlinesTrain)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

## testing the model

```python
test_dataset = countvector.transform(headlinesTest)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
